```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('sales_data.csv',encoding='ISO-8859-1')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.head()
```

```python
d=({'F':'Female','M':'Male'})
df['Gender']=df['Gender'].map(d)
df.head()
```

```python
df['Marital_Status']=df['Marital_Status'].map({0:'Unmarried',1:'Married'})
df.head()
```

```python
df.drop(['Age'],axis=1,inplace=True)
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
#  Gender
a=sns.countplot(x='Gender',data=df,hue='Gender')
for i in a.containers:
  a.bar_label(i)
```

```python
b=df.groupby('Gender')['Amount'].sum()
b=pd.DataFrame(b)
b
```

```python
b=df.groupby('Gender')['Amount'].sum()
b=pd.DataFrame(b)
a=sns.barplot(x='Gender',y='Amount',data=b)
for i in a.containers:
  a.bar_label(i)
```

Based on above plot female spend more money for shopping
